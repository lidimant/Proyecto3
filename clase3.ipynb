```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()

URL = "https://eurekabike.com/"

driver.get(URL)
time.sleep(20)  # esperar que cargue la página

# listas donde guardaremos los datos
nombres = []
precios = []
links = []

# buscar productos
productos = driver.find_elements(By.XPATH, "//a[contains(@href,'bike')]")

for p in productos:
    
    # nombre
    try:
        nombre = p.text
    except:
        nombre = None
    
    # enlace
    try:
        link = p.get_attribute("href")
    except:
        link = None
    
    nombres.append(nombre)
    links.append(link)
    
    # precio (si aparece cerca del producto)
    try:
        precio = p.find_element(By.XPATH, "..").text
    except:
        precio = None
    
    precios.append(precio)

# crear dataframe
df = pd.DataFrame({
    "bicicleta": nombres,
    "precio": precios,
    "link": links
})

# limpiar filas vacías
df = df[df["bicicleta"] != ""]

# guardar csv
df.to_csv("bicicletas.csv", index=False, encoding="utf-8")

print("Archivo bicicletas.csv creado correctamente")

driver.quit()
```
